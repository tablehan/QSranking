```python
#顯示檔案基本資料

def infodf(df):
    re="行數"+str(df.iloc[:,0].size)+"\n"
    re=re+"列數"+str(df.columns.size)+"\n"
    return re
```

```python
def gethtml(url2):
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.75 Safari/537.36"
    }
    #獲取資料
    req = requests.get(url=url2, headers=headers)
    req.encoding = "utf-8"
    return req.text
```

```python
def findfwci(text):
    a=text.find('fwciValue')
    if a=="-1":
        return "NAN"
    fwci=text[a:a+18]
    fwci
   
    return fwci
```
